```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
import re


# Set style for better looking plots
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# Load 2024 data files

host_dict = pd.read_csv('../../data/radio_programs_annotated-ai.csv')

p3_2024 = pd.read_csv('../../data/p3_2024.csv')
p4_2024 = pd.read_csv('../../data/p4_2024.csv')
p6_2024 = pd.read_csv('../../data/p6_2024.csv')
p3_2025 = pd.read_csv('../../data/p3_2025.csv')
p4_2025 = pd.read_csv('../../data/p4_2025.csv')
p6_2025 = pd.read_csv('../../data/p6_2025.csv')
```

```python
dictionary = {}

for index, row in host_dict.iterrows():
    episodeDescription = row['episodeDescription']
    host_name = row['selectedpred']
    if host_name == 'Error' or host_name == 'None':
        host_name = ''
    dictionary[episodeDescription] = host_name
    
```

```python
def update_dataframe_hosts_with_ai_annotation(df: pd.DataFrame) -> pd.DataFrame:
    df['hosts'] = df['hosts'].fillna(df['episodeDescription'].map(dictionary))
    return df
```

```python
p3_new_2024 = update_dataframe_hosts_with_ai_annotation(p3_2024)
p4_new_2024 = update_dataframe_hosts_with_ai_annotation(p4_2024)
p6_new_2024 = update_dataframe_hosts_with_ai_annotation(p6_2024)
p3_new_2025 = update_dataframe_hosts_with_ai_annotation(p3_2025)
p4_new_2025 = update_dataframe_hosts_with_ai_annotation(p4_2025)
p6_new_2025 = update_dataframe_hosts_with_ai_annotation(p6_2025)



p3_new_2024.to_csv("../../data/p3_2024.csv", index=False)
p4_new_2024.to_csv("../../data/p4_2024.csv", index=False)
p6_new_2024.to_csv("../../data/p6_2024.csv", index=False)

p3_new_2025.to_csv("../../data/p3_oct_2025.csv", index=False)
p4_new_2025.to_csv("../../data/p4_oct_2025.csv", index=False)
p6_new_2025.to_csv("../../data/p6_oct_2025.csv", index=False)

```
